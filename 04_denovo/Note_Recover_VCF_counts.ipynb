```python
import glob
import re
import pandas as pd
import os
```

```python
!pwd
```

```python
# Find files recursively em sub directories
files = glob.glob("00_denovo_populations/*_populations*/populations.log", recursive=True)  # Find in all sub directories
print(f"Find {len(files)} files")

```

```python
results = []

for file in files:
    try:
        with open(file, 'r') as f:
            for line in f:
                # Find specific line
                if 'variant sites remained' in line:
                    match = re.search(r'(\d+)\s*variant sites remained', line)
                    if match:
                        variant_sites = int(match.group(1))
                        results.append({
                            'File': file,
                            'Directory': os.path.dirname(file),
                            'Variant_Sites': variant_sites
                        })
                    break  # Stop after find it
                    
    except Exception as e:
        print(f"Error em {file}: {e}")

df = pd.DataFrame(results)
print(f"✅ Extracted {len(df)} values")
display(df.head())
```

```python
df['Directory'] = df['Directory'].str.replace('00_denovo_populations/','', regex='False')
df.head()
```

```python
df[["Group", "Missing"]] = df["Directory"].str.split("_populations_", n=1,expand=True)
df.head()
```

```python
df_ordem = df[['Group', 'Missing', 'Variant_Sites']]
df_ordem
```

```python
df_sorted = df_ordem.sort_values(['Group', 'Missing']).reset_index(drop=True)
```
