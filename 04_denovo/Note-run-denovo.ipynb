```python
#!/bin/bash

# Script mais simples baseado na sua estrutura exata

# Grupos e popmaps
groups=("alti" "grp" "phylo")
popmaps=("popmap-alti80" "popmap-grp41" "popmap-phylo29")

# Valores R e sufixos
r_values=("0.5" "0.9" "0.95")
r_suffixes=("50" "90" "95")

# Executar em paralelo
for i in "${!groups[@]}"; do
    group="${groups[i]}"
    popmap="${popmaps[i]}"
    
    for j in "${!r_values[@]}"; do
        r_value="${r_values[j]}"
        r_suffix="${r_suffixes[j]}"
        
        output_dir="${group}_populations_${r_suffix}"
        
        echo "ðŸš€ Executando: $group -> $output_dir (R=$r_value)"
        
        populations -P "$group"/ -O "$output_dir" -M "$popmap" -R "$r_value" \
            --min-maf 0.05 --write-single-snp --vcf --fstats -t 16 &
    done
done

# Esperar todos os processos
wait
echo "âœ… Todos os processos completados!"
```
