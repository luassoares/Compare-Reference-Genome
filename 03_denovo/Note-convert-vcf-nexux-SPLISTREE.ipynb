# Create the nexus from vcf to SPLITSTREE

```python
import os

# Create output folder
output_folder = "04_SPLITSTREE"
os.makedirs(output_folder, exist_ok=True)

# Read the list
with open('vcf_prefixes.txt', 'r') as f:
    prefixes = [line.strip() for line in f.readlines()]

print(f"Found {len(prefixes)} prefixes to process")

# Process using IPython magic
for prefix in prefixes:
    vcf_file = f"01_vcfs/{prefix}.vcf"
    output = "04_SPLITSTREE/"
    # Check if VCF file exists
    if not os.path.exists(vcf_file):
        print(f" VCF file not found: {vcf_file}")
        continue
    
    print(f"Processing: {prefix}")
    
    # Use IPython magic to run shell command
    !python3 ~/programs/vcf2phylip/vcf2phylip.py --input {vcf_file} --nexus --output-folder {output}
    
    print(f"Completed: {prefix}\n")

```
