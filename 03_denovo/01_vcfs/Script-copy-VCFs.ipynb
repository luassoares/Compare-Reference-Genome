# Extract vcf from denovo analysis

```python
import os
import shutil
import re
```

Pattern to extract groups and parameters from the folder name

```python
pattern = r'(.+)_populations_(\d+)'
```

```python
copied_files = 0
```

```python
# Destination folder
dest_dir = '01_vcfs'

print("ğŸ” Searching for VCF files...")

# Search only in folders that follow the pattern
for item in os.listdir('.'):
    if os.path.isdir(item) and '_populations_' in item:
        match = re.match(pattern, item)
        if match:
            group = match.group(1)
            param = match.group(2)
            
            # Search for VCF file
            for root, dirs, files in os.walk(item):
                for file in files:
                    if file.endswith('.vcf'):
                        vcf_path = os.path.join(root, file)
                        new_name = f"{group}_{param}.vcf"
                        dest_path = os.path.join(dest_dir, new_name)
                        
                        print(f"ğŸ“ Copying: {vcf_path}")
                        print(f"   â†ªï¸  To: {new_name}")
                        
                        shutil.copy2(vcf_path, dest_path)
                        copied_files += 1
                        break  # Exit after finding first VCF
                break  # Exit os.walk after finding

print(f"\nâœ… Process completed!")
print(f"ğŸ“Š Files copied: {copied_files}")
print(f"ğŸ“ Folder: {dest_dir}/")

# List result
print(f"\nğŸ“‹ Content of {dest_dir}/:")
for file in sorted(os.listdir(dest_dir)):
    size = os.path.getsize(os.path.join(dest_dir, file))
    print(f"  - {file} ({size:,} bytes)")
```
