```python

import numpy as np # array processing
import pandas as pd # table wrangling
from IPython.display import display
```

```python
popmap_grp = pd.read_csv('/data/users/sousal/03_compare_methods/popmap-grp41.txt', sep="\t")
popmap_alt = pd.read_csv('/data/users/sousal/03_compare_methods/popmap-alt80.txt', sep="\t")
popmap_phy = pd.read_csv('/data/users/sousal/03_compare_methods/popmap-phylo29.txt', sep="\t")

display(popmap_alt.head(), popmap_grp.head(), popmap_phy.head())
```

```python
popmap_grp['Sample']= popmap_grp['Sample'].astype(str)
popmap_alt['Sample']= popmap_alt['Sample'].astype(str)
```

```python
#Open the mulitqc file from the fastqc stats
fastqc_raw_dart1 = pd.read_csv('/data/users/sousal/01_RAW/RAW_DART_alti_cali_sche2020/fastqc/multiqc_data/multiqc_fastqc.txt', sep='\t')
fastqc_raw_dart2 = pd.read_csv('/data/users/sousal/01_RAW/RAW_dart_DPetu22-7327/fastqc_raw/multiqc_data/multiqc_fastqc.txt', sep='\t')
fastqc_raw_dart_calib_merged = pd.read_csv('/data/users/sousal/03_compare_methods/02_calib_merged_raw/fastqc/multiqc_data/multiqc_fastqc.txt', sep='\t')
```

```python
display(fastqc_raw_dart1.head(), fastqc_raw_dart2.head(), fastqc_raw_dart_calib_merged.head())
```

```python
#Remove .FASTQ from the Sample column
fastqc_raw_dart1['Sample'] = fastqc_raw_dart1['Sample'].str.replace('.FASTQ','', regex='False')
fastqc_raw_dart2['Sample'] = fastqc_raw_dart2['Sample'].str.replace('.FASTQ','', regex='False')
```

```python
grp = pd.merge(popmap_grp, 
				fastqc_raw_dart2[['Sample', 'Total Sequences','Sequence length']],
				on = 'Sample',
				how= 'left')

alt = pd.merge(popmap_alt, 
				fastqc_raw_dart1[['Sample', 'Total Sequences','Sequence length']],
				on = 'Sample',
				how= 'left')
				
phy = pd.merge(popmap_phy, 
				fastqc_raw_dart_calib_merged[['Sample', 'Total Sequences','Sequence length']],
				on = 'Sample',
				how= 'left')
```

```python
display(grp.head(),alt.head(),phy.head())
```

```python
alt['Group'] = 'ALT'
grp['Group'] = 'GRP' 
phy['Group'] = 'PHY'

df = pd.concat([grp,alt,phy]).reset_index(drop=True)
display(df)
```

```python
df = df.rename(
    columns={'Total Sequences': 'Raw Total_Sequences',
			'Sequence length': 'Raw Sequence length'}
)
df.head()
```

```python
fastqc_filtered_l60 = pd.read_csv('/data/users/sousal/03_compare_methods/02_filtered-adaptor-rmv-q30-l60-L60/fastqc/multiqc_data_1/multiqc_fastqc.txt', sep='\t')
```

```python
df_GRP = df[df['Group']== 'GRP'].reset_index(drop=True)
df_GRP.head()
```

```python
fastqc_filtered_l100 = pd.read_csv('/data/users/sousal/03_compare_methods/02_filtered-adaptor-rmv-q30-l100/fastqc/multiqc_data/multiqc_fastqc.txt', sep='\t')
fastqc_filtered_l100['Sample']= fastqc_filtered_l100['Sample'].astype(str)
```

```python
df_GRP.info()
```

```python
df_GRP_2 = pd.merge(df_GRP,
							 fastqc_filtered_l100[['Sample','Total Sequences', 'Sequence length']],
							 on= 'Sample',
							 how='left')
df_GRP_2.head()
```

```python
df_ALT_PHY = df[df['Group'] != 'GRP'].reset_index(drop=True)
df_ALT_PHY.head()
```

```python
fastqc_filtered = pd.read_csv('/data/users/sousal/03_compare_methods/02_filtered-adaptor-rmv-q30-l60-L60/fastqc/multiqc_data_1/multiqc_fastqc.txt', sep='\t')
```

```python
df_ALT_PHY_2 = pd.merge(df_GRP,
							 fastqc_filtered[['Sample','Total Sequences', 'Sequence length']],
							 on= 'Sample',
							 how='left')
df_GRP_2.head()
```

```python
df_2 = pd.concat([df_GRP_2, df_ALT_PHY_2]).reset_index(drop=True)
df_2.head()
```

```python
df_ordem=['Group', 'Sample', 'Pop', 'Raw Total_Sequences','Raw Sequence length','Total Sequences','Sequence length']
df_2 = df_2[df_ordem]
df_2 = df_2.rename(columns={'Total Sequences': 'Filtered Reads',
			'Sequence length': 'Filtered Sequence length (pb)'})
df_2.head()
```

```python
df_2.to_csv("Table-RAw-filtered-reads-csv", index=False)
```
